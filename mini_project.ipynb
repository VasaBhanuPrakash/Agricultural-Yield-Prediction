```python
from sklearn.preprocessing import StandardScaler,FunctionTransformer,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
def logtransformer(x):
    return np.log1p(x)
```

```python
features=['Area','Production','Annual_Rainfall','Fertilizer','Pesticide']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('log', FunctionTransformer(logtransformer, validate=True)),
            ('scaler', StandardScaler())
        ]),features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Crop', 'Season', 'State'])
    ])
```

```python
x=df.drop('Yield',axis=1)
y=df.Yield
```

```python
x_processed=preprocessor.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,accuracy_score,confusion_matrix,classification_report
X_train, X_test, y_train, y_test = train_test_split(x_processed, y, test_size=0.2, random_state=0)
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("mse is:",mean_squared_error(y_test,y_pred))
print("mae is:",mean_absolute_error(y_test,y_pred))
print("r2_score:",r2_score(y_test,y_pred))
y_pred=[1 if i>df.Yield.mean() else 0 for i in y_pred]
y_test=[1 if i>df.Yield.mean() else 0 for i in y_test]
print("accuracy is:",accuracy_score(y_test,y_pred))
print("Confusion Matrix:",confusion_matrix(y_test,y_pred))
print("Classsification Report:",classification_report(y_test,y_pred))
```
